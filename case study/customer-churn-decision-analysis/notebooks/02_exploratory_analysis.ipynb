```python
import pandas as pd

data = pd.read_csv("../data/synthetic_customer_churn.csv")
```

```python
data.info()
data.describe()
data['churn'].value_counts(normalize=True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.boxplot(x='churn', y='tenure_months', data=data)
plt.title("Tenure vs Churn")
plt.show()

sns.boxplot(x='churn', y='complaints', data=data)
plt.title("Complaints vs Churn")
plt.show()
```
